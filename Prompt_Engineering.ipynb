````python
from collections import Counter

def count_words_and_flag_difference(text):
    # Tokenize the string into words (assuming words are separated by spaces)
    words = text.split()

    # Count the occurrences of each word
    word_counts = Counter(words)

    # Print the word counts
    print("Word Counts:", word_counts)

    # Flag if the difference between the counts of any two words is less than 2
    words_list = list(word_counts.values())
    flag = False
    for i in range(len(words_list)):
        for j in range(i + 1, len(words_list)):
            if abs(words_list[i] - words_list[j]) < 2:
                flag = True
                break
        if flag:
            break

    # Return the word counts and the flag
    return word_counts, flag

# Example usage
text = "apple banana apple orange banana apple"
word_counts, flag = count_words_and_flag_difference(text)
print("Flag:", flag)
````
